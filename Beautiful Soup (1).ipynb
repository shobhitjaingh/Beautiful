Write a python program to extract information about the local weather from the National 
Weather Service website of USA, https://www.weather.gov/ for the city, San 
Francisco. You need to extract data about 7 day extended forecast display for the city. 
The data should include period, short description, temperature and description.

```python
weather=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJb7x7UzZPY")
weather
```

```python
weather_soup=BeautifulSoup(weather.content)
weather_soup
```

```python
report=weather_soup.find_all("li" ,class_="forecast-tombstone")
report
```

```python
Period=[]
Short_Desc=[]
HTemp=[]
LTemp=[]
for i in report:
    period=weather_soup.find_all("p" ,class_="period-name")
    
    short_desc=weather_soup.find_all("p",class_="short-desc")
    HTemperature=weather_soup.find_all("p",class_="temp temp-high")
  
    LTemperature=weather_soup.find_all("p",class_="temp temp-low")
    
for j in period:
        Period.append(j.text)    
for j in short_desc:
        Short_Desc.append(j.text) 
for j in HTemperature:    
        HTemp.append(j.text)  
for j in LTemperature:
        LTemp.append(j.text)        
        
    
  
   
print(Period)
print(Short_Desc)
print(HTemp)
print(LTemp)
```

```python
description=weather_soup.find_all("div",class_="col-sm-10 forecast-text")
description
```

```python
DESC=[]
for i in description:
    DESC.append(i.text)
    
DESC    
```

```python
def insert(a, b, positions):
    # reorder b and positions so that positions are in increasing order
    positions, b = zip(*sorted(zip(positions, b))) 

    out = []
    a_idx = 0
    it_b = iter(b)
    for pos in positions:
        slice_length = pos - len(out)
        out.extend(a[a_idx:a_idx + slice_length])
        out.append(next(it_b))
        a_idx += slice_length
    out.extend(a[a_idx:])
    return out
a=HTemp
b=LTemp
pos=[1,3,5,7]
TEMP=insert(HTemp,LTemp,pos)
TEMP
```

```python
import pandas as pd
data=list(zip(Period,Short_Desc,TEMP,DESC))
df=pd.DataFrame(data,columns=["PERIOD","SHORT DESCRIPTION","TEMPERATURE","DESCRIPTION"])
df
```
